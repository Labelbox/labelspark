# _**Exporting Labelbox Data to Databricks with LabelSpark**_

Install LabelSpark

```
%pip install labelspark -q
```

```
import labelspark as ls
```

Provide a Labelbox API Key and a Labelbox Project ID

```
api_key = "" # Labelbox API key
project_id = "" # Labelbox project ID
```

Initiate LabelSpark Client

```
client = ls.Client(api_key)
```

Export labels (and metadata), create a Spark DataFrame

```
df = client.export_to_table(
    project=project_id, 
    include_performance=True, # Will include columns that pertain to labeling / review performance
    include_agreement=True, # Will include agreement score column
    include_metadata=True, # Will include columns that pertain to metadata
    mask_method="png", # How to export masks from Labelbox - can be as a URL, PNG byte representation, or a numPy array representation ("url", "png", "array")
    verbose=True
)
```

View your Spark DataFrame

```
SAMPLE_TABLE = "labelspark_export_demo_test"
df.registerTempTable(SAMPLE_TABLE)
display(sqlContext.sql(f"select * from {SAMPLE_TABLE} LIMIT 5"))
```
